```python
# =============================================================================
# Downloading the up-to-date info
# =============================================================================

headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}

url = 'https://finance.yahoo.com/cryptocurrencies/'

# getting the live page
def get_page_contents():
    response = requests.get(url , headers = headers)
    return response.content

def convert_to_soup(content):
    return BeautifulSoup(content, features="html.parser")

def get_pattern(soup):
    global crypto_json
    pattern = re.compile(r'\s--\sData\s--\s')

    script_data = soup.find('script', text = pattern).contents[0]

    start = script_data.find("context")-2
    
    json_data = json.loads(script_data[start:-12])

    crypto_json = json_data['context']['dispatcher']['stores']['ScreenerResultsStore']['results']['rows']

def get_df(json_data):
    
    global df_cryptolist
    df_cryptolist = pd.json_normalize(crypto_json)
    
    # creating a dataset with the right columns and correct column names
    df_cryptolist = pd.DataFrame({'Symbol': df_cryptolist['symbol'],
                   'Name': df_cryptolist['shortName'],
                   'Price (Intraday)': df_cryptolist['regularMarketPrice.fmt']})
    
# Run live Website
def main():
    global df_cryptolist
    contents = get_page_contents()
    soup = convert_to_soup(contents)
    get_pattern(soup)
    get_df(crypto_json)


if __name__ == '__main__':
    main()
```

```python
def get_prices():

    prices = []
    crypto_name = []
    for crypto in cryptolist:
        price = str(df_cryptolist[df_cryptolist['Symbol'].str.contains(crypto)].iloc[:,2]).split(' ')[4].split('\n')[0]
        price = float(price.replace(",",""))
        prices.append(price)
        name= str(df_cryptolist[df_cryptolist['Symbol'].str.contains(crypto)].iloc[:,1]).split(' ')[4]
        crypto_name.append(name)
        
   # putting the column with prices in
    df_invest['price'] = prices
    df_invest['name'] = crypto_name
    
    # finding out how much is each investment
    df_invest['worth'] = df_invest['price'] * df_invest['amount'] 
```

```python
def print_table():
    #print a table
    print('============================================================')
    #total investment = summing the worth column
    print("The whole portfolio is worth ${:,.2f}".format(df_invest['worth'].sum()))
    print('============================================================')
    # go through each cryptocurrency and printing the amount, the crypto and its worth
    for index, row in df_invest.iterrows():
        print(row['amount'], row['crypto'], 'is worth ${:,.2f}'.format(row['worth']))
```

```python
# =============================================================================
# Asking the user: How much crypto have you got?
# =============================================================================
items = []
cryptolist = []
ask_again = ''
while ask_again != 'n':
    crypto = input('Please write the cryptocurrency symbol: ').upper()
    amount = float(input('Please write the amount of {} you have: '.format(crypto)))
    bought_price = float(input('Please write the price of each unit of {} that you bought it for: '.format(crypto)))
    spent = float(input('Please write down how much you spent on {}: '.format(crypto)))
    crypto_name = crypto +'-USD'
    cryptolist.append(crypto_name)
    item = [crypto, crypto_name, amount, bought_price, spent]
    items.append(item)
    ask_again = input('Would you add another cryptocurrency (y/n)? ')
    


#creatubg a dataframe
df_invest= pd.DataFrame(items,  columns = ['crypto',
                                           'crypto_name',
                                           'amount',
                                           'bought_price',
                                           'spent'])

get_prices()
print_table()

# other important variables
df_invest['initial_invesment'] = df_invest['bought_price'] * df_invest['amount'] 

df_invest['profit'] = df_invest['worth'] - df_invest['spent'] 

df_invest.head()

```

```python
# =============================================================================
# Downloading the necesary datasets
# =============================================================================
# creating a setting for download
start = "2020-08-01"
end = dt.datetime.now()
short_sma = 20
long_sma = 50


# creating a dataset for each cryptocurrency that the user has put in 
df = {}
for crypto in cryptolist :
    data = yf.download(crypto, start, end,interval = '1d')
    df[crypto] = pd.DataFrame(data.dropna(), columns = ['Open', 'High','Low','Close', 'Adj Close', 'Volume'])
    df[crypto]['short_SMA'] = df[crypto].iloc[:,1].rolling(window = short_sma).mean()
    df[crypto]['long_SMA'] = df[crypto].iloc[:,1].rolling(window = long_sma).mean()
    df[crypto].index = pd.to_datetime(df[crypto].index)
    df[crypto] = df[crypto].asfreq('D')
#    print('Nan in each columns before Backfill for' , crypto)
#    print(df[crypto] .isna().sum())
    df[crypto] = df[crypto].bfill()
#    print(df[crypto].head())
```
